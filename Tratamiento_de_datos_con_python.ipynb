```python
import pandas as pd

# Cargar el archivo CSV
# Reemplaza 'nombre_archivo.csv' con el nombre real de tu archivo CSV
data = pd.read_csv('/content/datos_normalizados.csv')

# Función para buscar los datos por matrícula
def buscar_por_matricula(matricula):
    usuario = data[data['matricula'] == matricula]
    return usuario if not usuario.empty else None

# Solicitar la matrícula al usuario
matricula_buscar = int(input("Ingrese la matrícula a buscar: "))

# Buscar los datos del usuario
datos_usuario = buscar_por_matricula(matricula_buscar)

# Mostrar los datos si se encontraron, de lo contrario, indicar que no se encontró la matrícula
if datos_usuario is not None:
    print("Datos del usuario con matrícula", matricula_buscar)
    print(datos_usuario)
else:
    print("La matrícula ingresada no se encontró en el archivo.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
data = pd.read_csv('datos_normalizados.csv')

# Crear subplots para mostrar gráficos de las variables
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

# Graficar la distribución de género
data['genero'].value_counts().plot(kind='bar', ax=axs[0, 0], color='salmon')
axs[0, 0].set_title('Distribución de Género')
axs[0, 0].set_xlabel('Género')
axs[0, 0].set_ylabel('Cantidad')
axs[0, 0].grid(axis='y')

# Graficar la distribución de edades
axs[0, 1].hist(data['edad'], bins=10, color='skyblue')
axs[0, 1].set_title('Distribución de Edades')
axs[0, 1].set_xlabel('Edad')
axs[0, 1].set_ylabel('Frecuencia')
axs[0, 1].grid(True)

# Graficar la distribución de pesos
axs[0, 2].hist(data['peso'], bins=10, color='lightgreen')
axs[0, 2].set_title('Distribución de Pesos')
axs[0, 2].set_xlabel('Peso')
axs[0, 2].set_ylabel('Frecuencia')
axs[0, 2].grid(True)

# Graficar la distribución de estaturas
axs[1, 0].hist(data['estatura'], bins=10, color='lightcoral')
axs[1, 0].set_title('Distribución de Estaturas')
axs[1, 0].set_xlabel('Estatura')
axs[1, 0].set_ylabel('Frecuencia')
axs[1, 0].grid(True)

# Graficar un gráfico de pastel de confirmar-cronica
confirmar_cronica = data['confirmar-cronica'].value_counts()
axs[1, 1].pie(confirmar_cronica, labels=confirmar_cronica.index, autopct='%1.1f%%', colors=['lightblue', 'lightyellow'])
axs[1, 1].set_title('Porcentaje Confirmar-Cronica')
axs[1, 1].axis('equal')

# Graficar un gráfico de pastel de enfermedades
enfermedades = data['enfermedad'].value_counts()
axs[1, 2].pie(enfermedades, labels=enfermedades.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'])
axs[1, 2].set_title('Porcentaje de Enfermedades')
axs[1, 2].axis('equal')

# Graficar un gráfico de barras de familiar-enfermo
data['familiar-enfermo'].value_counts().plot(kind='bar', ax=axs[2, 0], color='lightgrey')
axs[2, 0].set_title('Familiar Enfermo')
axs[2, 0].set_xlabel('Familiar Enfermo')
axs[2, 0].set_ylabel('Cantidad')
axs[2, 0].grid(axis='y')

# Graficar un gráfico de barras de enfermedad-familiar
data['enfermedad-familiar'].value_counts().plot(kind='bar', ax=axs[2, 1], color='lightpink')
axs[2, 1].set_title('Enfermedad Familiar')
axs[2, 1].set_xlabel('Enfermedad Familiar')
axs[2, 1].set_ylabel('Cantidad')
axs[2, 1].grid(axis='y')

# Graficar un gráfico de barras de lugar-residencial
data['lugar-residencial'].value_counts().plot(kind='bar', ax=axs[2, 2], color='lightgreen')
axs[2, 2].set_title('Lugar Residencial')
axs[2, 2].set_xlabel('Lugar Residencial')
axs[2, 2].set_ylabel('Cantidad')
axs[2, 2].grid(axis='y')

plt.tight_layout()
plt.show()
```
